```julia
using Printf
using Base.Threads
using Distributed

# Função com carga computacional pesada para demonstrar benefício do paralelismo
function trabalho_pesado(i)
    x = 0.0
    # Loop interno intensivo - muitas operações matemáticas
    for k in 1:70000
        x += sin(i * k * 0.0001) * cos(i * k * 0.0001)
    end
    return x
end

# Versão série - processa uma tarefa por vez
function loop_serie()
    n = 10000
    resultados = zeros(n)

    # Loop sequencial tradicional
    for i in 1:n
        resultados[i] = trabalho_pesado(i)
    end

    return sum(resultados)
end

# Versão paralela - divide tarefas entre threads
function loop_paralelo()
    n = 10000
    resultados = zeros(n)

    # Macro @threads distribui automaticamente as iterações
    @threads for i in 1:n
        resultados[i] = trabalho_pesado(i)
    end

    return sum(resultados)
end

# Medição comparativa de desempenho
tempo_serie = @elapsed loop_serie()
tempo_paralelo = @elapsed loop_paralelo()

# Exibe resultados
@printf("Tempo série: %.6f\n", tempo_serie)
@printf("Tempo paralelo: %.6f\n", tempo_paralelo)
```
